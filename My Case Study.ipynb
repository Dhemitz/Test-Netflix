## 3. Data Preprocessing

### Director and cast

```python
movies.drop(['director', 'cast'], axis=1, inplace=True) # menghapus kolom director & cast
```

```python
movies.isnull().sum()
```

### Country

```python
movies['country'].mode()[0]
```

```python
movies['country'] = movies['country'].fillna(movies['country'].mode()[0]) # mengubah isi counry Null menjadi country yang sering muncul
```

```python
movies.isnull().sum()
```

### Rating

```python
null_rating = movies['rating'].isnull()
null_rating
```

```python
null_rating_dataframe = movies[null_rating]
null_rating_dataframe
```

```python
rating_replacements = {
    5989: 'TV-PG',
    6827: 'TV-14',
    7312: 'TV-MA',
    7537: 'MR'
}
```

```python
rating_replacements.items()
```

```python
for id, rate in rating_replacements.items():
  movies.loc[id, 'rating'] = rate
```

```python
movies.isnull().sum()
```

### Date Added

```python
movies['date_added'] = movies[['date_added']].replace(np.nan, 'Not Added')
```

```python
movies.isnull().sum()
```

### Duration

```python
null_duration = movies['duration'].isnull()
null_duration
```

```python
null_duration_dataframe = movies[null_duration]
null_duration_dataframe
```

```python
duration_replacements = {
    5541: '115 min',
    5794: '115 min',
    5813: '115 min'
}
```

```python
duration_replacements.items()
```

```python
for id, rate in duration_replacements.items():
  movies.loc[id, 'duration'] = rate
```

```python
movies.isnull().sum()
```

## 4. Data Visualization

```python
movies['release_year'].value_counts().index[0:15]
```

```python
plt.figure(figsize=(8, 6))
sns.set(style='whitegrid')
sns.countplot(y= 'release_year', data=movies, order=movies['release_year'].value_counts().index[0:15])
plt.title('Release Year', fontsize=12, fontweight='bold')
plt.xlabel('Total')
plt.ylabel('Year')
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=movies, palette='rocket')
plt.title('TV Show vs Movie', fontsize=12, fontweight='bold')
plt.xlabel('Type')
plt.ylabel('Total')
```

```python
plt.figure(figsize=(11, 8))
sns.countplot(x='rating', palette='rocket', data=movies, order=movies['rating'].value_counts().index[0:15])
```

```python
movies['release_month'] = movies['date_added'].apply(lambda x: x.lstrip().split(" ")[0])
movies.head()
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(x='release_month', data=movies, order=movies['release_month'].value_counts().index[0:12])
plt.xticks(rotation=45)
```
